```python
def bindec(b):
    b_inteiro = str(int(float(b)))
    b_fracao = ""
    d = 0.0
    calculo1 = ""
    calculo2 = ""
    
    for i in range(len(b)):
        if b[i] == ".":
            for x in range(i+1, len(b)):
                b_fracao += b[x]
            break
    
    b = b_inteiro + b_fracao
    
    for i in range(len(b)):
        if i==(len(b)-1):
            calculo1 += b[i] + " x 2^" + str((len(b_inteiro)-1)-i)
            calculo2 += str(int(b[i]) * (2**((len(b_inteiro)-1)-i)))
        else:
            calculo1 += b[i] + " x 2^" + str((len(b_inteiro)-1)-i) + " + "
            calculo2 += str(int(b[i]) * (2**((len(b_inteiro)-1)-i))) + " + "
        d += int(b[i]) * (2**((len(b_inteiro)-1)-i))
    print(calculo1)
    print(calculo2)
    print(" = ", d)
    return d
    

def decbin(d):
    d_inteiro = int(d)
    d_fracao = d - d_inteiro
    b_inteiro = ""
    b_inteiro2 = ""
    b_fracao = ""
    
    """Decimal Inteiro >>> Binário"""
    div = d_inteiro
    while div != 1:
        print(div, "/2 = ", int(div/2), " resto ", div % 2)
        b_inteiro += str(div % 2)
        div = int(div/2)
    b_inteiro += "1"

    for i in range(len(b_inteiro)-1, -1, -1):
        b_inteiro2 += b_inteiro[i]

    """Decimal Fracionário >>> Binário"""
    mult = d_fracao
    while mult != 1:
        print(mult, " x 2 = ", mult*2)
        b_fracao += str(int(mult*2))
        if ((mult*2) - int(mult*2)) == 0:
            break
        else:
            mult = (mult*2) - int(mult*2)
    
    return b_inteiro2 + "." + b_fracao 

escolha = int(input("\nDigite:\n1 para converter Binário em Decimal\n2 para converter Decimal em Binário\n"))
while escolha != 1 and escolha != 2:
    escolha = int(input("\nValor inválido!"
                        +"\nDigite:\n1 para converter Binário em Decimal\n2 para converter Decimal em Binário\n"))

if escolha == 1:
    teste = True
    while teste:
        binario = input("Digite o número binário: ")
        for i in range(len(binario)):
            if binario[i] != "0" and binario[i] != "1" and binario[i] != ".":
                teste = True
                print("Valor inválido!")
                break
            teste = False
    print("\nBinário: ", binario, "\nDecimal: ", bindec(binario))
else:
    decimal = float(input("\nDigite o número decimal: "))
    print("\nDecimal: ", decimal, "\nBinário: ", decbin(decimal))
    
```
