```python
def binoct(b):
    o = ""
    
    while (len(b) % 3) != 0:
        b = "0" + b
        
    for i in range(0, len(b), 3):
        print(b[i]+b[i+1]+b[i+2])
        o += str(int(b[i])*4 + int(b[i+1])*2 + int(b[i+2])*1)
        print(b[i] + " x 2^2 + " + b[i+1] + " x 2^1 + " + b[i+2] + " x 2^0 = "
              + str(int(b[i])*4) + " + " + str(int(b[i+1])*2) + " + "
              + str(int(b[i+2])*1) + " = " + o[len(o)-1])
    
    return o

def octbin(o):
    
    b = ""
    
    for i in range(len(o)):
        if o[i] == "0": b += "000"
        elif o[i] == "1": b += "001"
        elif o[i] == "2": b += "010"
        elif o[i] == "3": b += "011"
        elif o[i] == "4": b += "100"
        elif o[i] == "5": b += "101"
        elif o[i] == "6": b += "110"
        elif o[i] == "7": b += "111"
        print(o[i] + " = "+b[len(b)-3]+" + "+b[len(b)-2]+" + "+b[len(b)-1]
              +" = "+b[len(b)-3]+" x 2^2 + "+b[len(b)-2]+" x 2^1 + "
              +b[len(b)-1]+" x 2^0 --> "+b[len(b)-3]+b[len(b)-2]+b[len(b)-1])
    
    return str(int(b))

def hexletra(n):
    if n == 10:
        return "A"
    elif n == 11:
        return "B"
    elif n == 12:
        return "C"
    elif n == 13:
        return "D"
    elif n == 14:
        return "E"
    else: return "F"

def binhex(b):
    h = ""
    
    while True:
        if (len(b) % 4) == 0:
            break
        else:
            b = "0" + b
        
    for i in range(0, len(b), 4):
        print(b[i]+b[i+1]+b[i+2]+b[i+3])
        num = int(b[i])*8 + int(b[i+1])*4 + int(b[i+2])*2 + int(b[i+3])*1
        if num > 9:
            h += str(hexletra(num))
        else: h += str(num)
        print(b[i] + " x 2^3 + " + b[i+1] + " x 2^2 + " + b[i+2] + " x 2^1 + "
              + b[i+3] + " x 2^0 = "+ str(int(b[i])*8) + " + " + str(int(b[i+1])*4) 
              + " + " + str(int(b[i+2])*2) + " + " + str(int(b[i+3])*1)
              + " = " + h[len(h)-1])
    
    return h

def hexbin(h):
    
    b = ""
    
    for i in range(len(h)):
        if h[i] == "0": b += "0000"
        elif h[i] == "1": b += "0001"
        elif h[i] == "2": b += "0010"
        elif h[i] == "3": b += "0011"
        elif h[i] == "4": b += "0100"
        elif h[i] == "5": b += "0101"
        elif h[i] == "6": b += "0110"
        elif h[i] == "7": b += "0111"
        elif h[i] == "8": b += "1000"
        elif h[i] == "9": b += "1001"
        elif h[i] == "A": b += "1010"
        elif h[i] == "B": b += "1011"
        elif h[i] == "C": b += "1100"
        elif h[i] == "D": b += "1101"
        elif h[i] == "E": b += "1110"
        elif h[i] == "F": b += "1111"
        print(h[i] + " = "+b[len(b)-4]+" + "+b[len(b)-3]+" + "+b[len(b)-2]+" + "
              +b[len(b)-1]+" = "+b[len(b)-4]+" x 2^3 + "+b[len(b)-3]+" x 2^2 + "
              +b[len(b)-2]+" x 2^1 + "+b[len(b)-1]+" x 2^0 --> "
              +b[len(b)-4]+b[len(b)-3]+b[len(b)-2]+b[len(b)-1])
        
    return str(int(b))

def decbin(d):
    b1 = ""
    b2 = ""

    div = int(d)
    while div != 1:
        print(div, "/2 = ", int(div/2), " resto ", div % 2)
        b1 += str(div % 2)
        div = int(div/2)
    b1 += "1"

    for i in range(len(b1)-1, -1, -1):
        b2 += b1[i]
    
    print("Binário: "+ b2)
    return b2

def bindec(b):
    print("Binário: "+b)
    d = 0

    calculo1 = ""
    calculo2 = ""

    for i in range(len(b)):
        if i==(len(b)-1):
            calculo1 += b[i] + " x 2^" + str((len(b)-1)-i)
            calculo2 += str(int(b[i]) * (2**((len(b)-1)-i)))
        else:
            calculo1 += b[i] + " x 2^" + str((len(b)-1)-i) + " + "
            calculo2 += str(int(b[i]) * (2**((len(b)-1)-i))) + " + "

        d += int(b[i]) * (2**((len(b)-1)-i))

    print(calculo1)
    print(" = " + calculo2)
    print(" =", d)
    
    return d

escolha = int(input("Digite para converter:\n"
                    +"1 - Binário -> Octal\n"
                    +"2 - Octal -> Binário\n"
                    +"3 - Binário -> Hexadecimal\n"
                    +"4 - Hexadecimal -> Binário\n"
                    +"5 - Decimal -> Octal\n"
                    +"6 - Octal -> Decimal\n"
                    +"7 - Decimal -> Hexadecimal\n"
                    +"8 - Hexadecimal -> Decimal\n"))

while escolha < 1 or escolha > 8:
    escolha = int(input("Valor inválido!\nDigite para converter:\n"
                    +"1 - Binário -> Octal\n"
                    +"2 - Octal -> Binário\n"
                    +"3 - Binário -> Hexadecimal\n"
                    +"4 - Hexadecimal -> Binário\n"
                    +"5 - Decimal -> Octal\n"
                    +"6 - Octal -> Decimal\n"
                    +"7 - Decimal -> Hexadecimal\n"
                    +"8 - Hexadecimal -> Decimal\n"))

if escolha == 1:
    teste = True
    while teste:
        binario = input("Digite o número binário: ")
        for i in range(len(binario)):
            if binario[i] != "0" and binario[i] != "1":
                teste = True
                print("Valor inválido!")
                break
            teste = False
    print("")
    print("\nBinário: ", binario, "\nOctal: ", binoct(binario))
elif escolha == 2:
    octal = input("Digite o número octal: ")
    print("")
    print("\nOctal: ", octal, "\nBinário: ", octbin(octal))
elif escolha == 3:
    teste = True
    while teste:
        binario = input("Digite o número binário: ")
        for i in range(len(binario)):
            if binario[i] != "0" and binario[i] != "1":
                teste = True
                print("Valor inválido!")
                break
            teste = False
    print("")
    print("\nBinário: ", binario, "\nHexadecimal: ", binhex(binario))
elif escolha == 4:
    hexadecimal = input("Digite o número hexadecimal: ")
    print("")
    print("\nHexadecimal: ", hexadecimal, "\nBinário: ", hexbin(hexadecimal))
elif escolha == 5:
    decimal = input("Digite o número decimal: ")
    print("")
    print("\nDecimal: ", decimal, "\nOctal: ", binoct(decbin(decimal)))
elif escolha == 6:
    octal = input("Digite o número octal: ")
    print("")
    print("\nOctal: ", octal, "\nDecimal: ", bindec(octbin(octal)))
elif escolha == 7:
    decimal = input("Digite o número decimal: ")
    print("")
    print("\nDecimal: ", decimal, "\nHexadecimal: ", binhex(decbin(decimal)))
elif escolha == 8:
    hexadecimal = input("Digite o número hexadecimal: ")
    print("")
    print("\nHexadecimal: ", hexadecimal, "\nDecimal: ", bindec(hexbin(hexadecimal)))
     
```
